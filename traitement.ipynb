# Etape 2 : Nettoyage des données

```python
data_a.head(100)
```

```python
data_b.head(100)
```

```python
import re

def reformat_phone_number(phone_number):
    
    try:
        if (pd.isna(phone_number)):
            return None
    
        if (phone_number.startswith("33")):
            phone_number = "+" + phone_number
    
        elif re.match(r'^0[1-9]', phone_number):
            phone_number = "+33 " + phone_number[1:]
        
        return re.sub(r'[() -]', '', phone_number)
    except:
        return None

data_a['Telephone'] = data_a['Telephone'].apply(reformat_phone_number)
data_b['Telephone'] = data_b['Telephone'].apply(reformat_phone_number)
```

```python
remplacements = {'[éè]': 'e', 'ô': 'o', '-': ' ', '_': ' ', '/': ' / ', "\.": " "}

data_a['Nom'] = data_a['Nom'].str.lower()
data_a['Nom'] = data_a['Nom'].replace(remplacements, regex=True)
data_b['Nom'] = data_b['Nom'].str.lower()
data_b['Nom'] = data_b['Nom'].replace(remplacements, regex=True)
```

```python
remplacements = {'\.0': ''}

data_a['Zipcode'] = data_a['Zipcode'].replace(remplacements, regex=True)
data_b['Zipcode'] = data_b['Zipcode'].replace(remplacements, regex=True)
```

```python
import numpy as np

data_a = data_a.replace(to_replace=[np.nan], value=None)
data_b = data_b.replace(to_replace=[np.nan], value=None)
```

# Etape 3 : Définir un plan pour le matching

## Probleme : beaucoup trop de candidat
Si on choisit de comparer toutes les lignes avec toutes les lignes, on va avoir 10000 * 10000 = 100 millions de combinaison
Rien d'impossible, mais pas viable à une autre echelle

```python
indexer = recordlinkage.Index()
indexer.full() # Cette ligne permet de comparer toutes les lignes avec toutes les lignes

candidate_links = indexer.index(data_a, data_b)

compare = recordlinkage.Compare()
compare.string('Nom', 'Nom', method='jarowinkler', threshold=0.8)
compare.exact('Telephone', 'Telephone')
compare.exact('Zipcode', 'Zipcode')
compare.string('Site', 'Site', method='jarowinkler', threshold=0.9);

# resultats = compare.compute(candidate_links, data_a, data_b) # Trop long pour être executé (100 millions d'occurance)
# print(resultats)
```

## Proposition de methode pour affiner la liste des candidats
- Comparer 2 lignes qui on la même proprité ID (comme vu au dessus avec des ID similaires)
- Extraite une carateristique du nom de l'entreprise pour separer les candidats (ex: lettre la plus presente dans le nom)
- Domaine du site internet identique
- Code postal identique

### Id identique (A)

```python
indexer = recordlinkage.Index()
indexer.block("Id") # Cette ligne permet de comparer toutes les lignes avec toutes les lignes

candidate_links = indexer.index(data_a, data_b)

compare = recordlinkage.Compare()
compare.exact('Nom', 'Nom')
compare.string('Nom', 'Nom', method='jarowinkler', threshold=0.8)

resultats = compare.compute(candidate_links, data_a, data_b)
print("Nombre de combinaison testé = ", len(resultats))
```

```python
resultats.head(5)
```

**Toutes les lignes ont un nom identique**

```python
resultats_a = resultats[resultats[[0]].sum(axis=1) >= 1]
resultats_a["match_a"] = 2 # 2 = total, 1 = partiel

print("Nombre de match trouvé = ", len(resultats_a))
```

**Conclusion : On voit bien que chaque ID similaire mene a un matching (au moins un match ou nom similaire)**

### Zipcode identique (B)
Si le nom et le zipcode sont identique, alors c'est un matching exact

```python
indexer = recordlinkage.Index()
indexer.block("Zipcode") # Cette ligne permet de comparer toutes les lignes avec toutes les lignes

candidate_links = indexer.index(data_a, data_b)

compare = recordlinkage.Compare()
compare.string('Nom', 'Nom', method='jarowinkler', threshold=0.8)

resultats = compare.compute(candidate_links, data_a[data_a["Zipcode"] != None], data_b[data_b["Zipcode"] != None])
print("Nombre de combinaison testé = ", len(resultats))
```

```python
resultats_b = resultats[resultats[0] == 1] # On recherche les matchs par nom trouvé
resultats_b["match_b"] = 2 # 2 = total, 1 = partiel

print("Nombre de match par nom = ", len(resultats_b))
```

### Lettre la plus presente dans le nom de l'entreprise (C)

```python
def lettre_la_plus_frequente(texte):    
    chaine = texte.replace(" ", "").lower()
    
    if not chaine:
        return None  # Gestion des cas où la chaîne est vide

    lettre_plus_presente = max(set(chaine), key=chaine.count)
    return lettre_plus_presente

data_a['LettrePlusFrequente'] = data_a['Nom'].apply(lettre_la_plus_frequente)
data_b['LettrePlusFrequente'] = data_b['Nom'].apply(lettre_la_plus_frequente)
```

```python
data_a.head(10)
```

```python
indexer = recordlinkage.Index()
indexer.block("LettrePlusFrequente") # Cette ligne permet de comparer toutes les lignes avec toutes les lignes

candidate_links = indexer.index(data_a, data_b)

compare = recordlinkage.Compare()
compare.string('Nom', 'Nom', method='jarowinkler', threshold=0.9)

resultats = compare.compute(candidate_links, data_a, data_b)
print("Nombre de combinaison testé = ", len(resultats))
```

```python
resultats_c = resultats[resultats[0] == 1]
resultats_c["match_c"] = 1 # Ici, on definit en match partiel car seul le nom match. 
# Si c'est en réalité un match exact, on viendra trouver cette information dans les resultats A ou B
```

```python
print("Match avec le même nom d'entreprise = ", len(resultats_c))
```

```python
resultats_c.head(10)
```

# Etape 4 : Génération du fichier
On va fusionner nos resultats A, B et C, supprimer les diplications et en faire un CSV

```python
print(len(resultats_a))
resultats_a = resultats_a[["match_a"]]
resultats_a.head()
```

```python
print(len(resultats_b))
resultats_b = resultats_b[["match_b"]]
resultats_b.head()
```

```python
print(len(resultats_c))
resultats_c = resultats_c[["match_c"]]
resultats_c.head()
```

```python
total = len(resultats_a) + len(resultats_b) + len(resultats_c)
print("Nombre de ligne total = ", total)
```

```python
merged_df = resultats_a[["match_a"]].merge(resultats_b[["match_b"]], left_index=True, right_index=True, how='outer')
merged_df = merged_df.merge(resultats_c[["match_c"]], left_index=True, right_index=True, how='outer')

merged_df["match"] = merged_df.max(axis=1)

print("Nombre de ligne apres merge = ", len(merged_df))
```

```python
merged_df.head(20)
```

On peut voir ici quelle methode de matcing a détécté le match. On voit que la derniere methode (C) à partir d'une lettre a donné beaucoup de ligne.

**Super, on trouve des lignes avec un nom de boutique identique**
Malheuresement, il n'est pas possible de savoir si il s'agit d'une chaine ou bien de 2 entreprises differentes qui ont utilisés le même nom

**On vient transformer le chiffre en TAG**

```python
def set_tag(row):
    if row["match"] == 2:
        return "exact"
    return "partiel"

merged_df["match"] = merged_df.max(axis=1)
merged_df["tag"] = merged_df.apply(set_tag, axis=1)

match_final = merged_df[["tag"]]
match_final = match_final.reset_index()
match_final = match_final.rename(columns={'level_0': 'index_a', 'level_1': 'index_b'})
match_final.head()
```

```python
# Sauvegarde dans un fichier resultat
print("Nombre de matching trouvé = ", len(match_final))
match_final.to_csv("matching.csv", index=False)
```
